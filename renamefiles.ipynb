# RENAMING FILE NAMES FROM NUM TO THEIR ACTUAL CHAR NAME

Initially the dataset we had were not named according to their character name but were assigned a number corresponding to the character name.

<b>For example,</b> directory with name 1 inside directory vowels corresponds to character 'a'.

To make programming easier, renaming of directory to their actual character name is done.

## Renaming consonants directory

```python
#The consonants are present from 34-69
numconsonants=np.array(df[34:].class_) 
#numconsonants contain the number by which the file containing images of the corresponding devnagri char is sored in
print(numconsonants)
```

```python
devnagriconsonants=np.array(df[34:].devnagrilabel)
```

```python
labelconsonants =np.array(df[34:].label)
#labelconsonants give the actual name of the devnagri character 
labelconsonants
```

```python
#As we can see 'dha' is present 2X at index 13 and 23
#AT index 23 it should be bha
labelconsonants[23]='bha'
```

```python
#first we will change the dtype of numconsonants from object to int32
print(numconsonants.dtype)
numconsonants=numconsonants.astype('int32')
print(numconsonants.dtype)
```

```python
#Now we can access the labels simply by the numconsonants value
for x in numconsonants :
    print(str(x)+" "+labelconsonants[x-1])
```

```python
#Now we will rename the the consonant names to their respective labels 
l1=list(map(str,range(1,37)))
consonants=os.path.join(dir,'consonants')
for x in os.listdir(consonants):
    print(x)
    if x in l1:
        print("came here")
        print(labelconsonants[int(x)-1])
        os.rename(os.path.join(consonants,x),os.path.join(consonants,labelconsonants[int(x)-1]))
```
